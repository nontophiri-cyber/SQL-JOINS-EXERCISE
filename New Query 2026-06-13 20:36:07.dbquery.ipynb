```sql
QUESTION 01: Show every User who has subscription.
-------------------------------------------------------------------
SELECT u.user_id, u.user_name, s.start_date, s.subscriptio_id
FROM workspace.default.users_table u
INNER JOIN workspace.default.subscriptions_table s
ON u.user_id = s.user_id;
--------------------------------------------------------------------
Question 02: Show every subscription with its matching plan name and monthly price

SELECT s.subscriptio_id ,s.user_id , p.plan_name, p.monthly_price
FROM workspace.default.plans_table p
    INNER JOIN  workspace.default.subscriptions_table s
    ON p.plan_id = s.plan_id 
-----------------------------------------------------------------------
QUESTION 03:Show every viewing session that has a matching show. Include the show title and genre.
SELECT v.session_id, v.user_id, h.show_title, h.genre, v.watch_minutes
FROM workspace.default.shows_tables AS h
INNER JOIN workspace.default.viewing_sessions AS v
ON v.show_id = h.show_id
-------------------------------------------------------------------------
QUESTION 04: Show every viewing session with the user who watched it. Only show sessions with a matching user.
SELECT u.user_name, u.country, v.session_id, v.show_id, v.watch_minutes
FROM workspace.default.users_table u
INNER JOIN workspace.default.viewing_sessions v
ON u.user_id = v.user_id
--------------------------------------------------------------------------
QUESTION 05: Show users along with their subscriptions, the plan name, and the price. Use only users who have both a subscription and a valid plan.

SELECT u.user_name, u.country, p.plan_name, p.monthly_price, s.start_date
FROM workspace.default.users_table u
INNER JOIN  workspace.default.subscriptions_table s
ON u.user_id = s.user_id
INNER JOIN workspace.default.plans_table p
ON s.plan_id = p.plan_id
-----------------------------------------------------------------------
                     PART 04: LEFT JOIN
QUESTION 06: Show every user and any subscriptions they have. Users without subscriptions must still appear.
SELECT u.user_id, u.user_name, s.start_date, s.subscriptio_id
FROM workspace.default.users_table u
LEFT JOIN workspace.default.subscriptions_table s
ON u.user_id = s.user_id
------------------------------------------------------------------------
QUESTION 07: Show every plan and the subscriptions on it. Plans with no subscribers must still appear.
SELECT p.plan_id, p.plan_name, s.subscriptio_id, s.user_id
FROM workspace.default.plans_table p
LEFT JOIN workspace.default.subscriptions_table s
ON p.plan_id = s.plan_id
------------------------------------------------------------------------
QUESTION 08 : Show every show and any viewing sessions on it. Shows that were never watched must still appear.
SELECT h.show_id, h.show_title , v.session_id, v.watch_minutes
FROM workspace.default.shows_tables h
LEFT JOIN workspace.default.viewing_sessions v
ON h.show_id = v.show_id 
---------------------------------------------------------------------------
QUESTION 09: Show every viewing session and the user who watched it. Sessions referencing users that do not exist must still appear

SELECT v.session_id, v.show_id, v.watch_minutes, u.user_id, u.user_name
FROM workspace.default.viewing_sessions v
LEFT JOIN workspace.default.users_table u
ON v.user_id = u.user_id 
----------------------------------------------------------------------------
QUESTION 10: Show every user, the plan they are on (if any), and the monthly price. Users without a subscription must still appear.

SELECT u.user_name, u.country, p.plan_name, p.monthly_price
FROM workspace.default.users_table u 
LEFT JOIN  workspace.default.subscriptions_table s
ON u.user_id = s.user_id 
LEFT JOIN workspace.default.plans_table p 
ON s.plan_id = p.plan_id
----------------------------------------------------------------------------
                       09. PART C FULL OUTER JOIN
QUESTION 11: Show every user and every subscription, including users without subscriptions AND subscriptions referencing users that do not exist
```
